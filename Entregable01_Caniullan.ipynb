# CARGAR DATA USANDO API

###Extracción de la data

```python
# importar librerias
import requests
import json
import pandas as pd
import psycopg2
```

```python
# conectar a base de datos via API
response = requests.get(url="https://www.datos.gov.co/resource/p39t-z7qc.json")
response
```

```python
# revisar data en formato txt
response.text
```

### Transformar data

```python
# transformar data de txt a json y guardar en df1
df1 = json.loads(response.text)
df1
```

```python
# transformar df1 de json a df2 dataframe
df2=pd.DataFrame(df1)
df2.head()
```

### EDA

```python
# revisar dimensiones de df2
df2.shape
```

```python
# contar la cantidad de valores NA en cada variable
df2.isna().sum()
```

```python
# cantidad de valores unicos en la variable 'personal'
df2['personal'].value_counts()
```

```python
# transformar de formato las variables
from datetime import date
df2['matricula'] = df2['matricula'].astype(str)
df2['organizacion'] = df2['organizacion'].astype(str)
df2['razon_social'] = df2['razon_social'].astype(str)
df2['identificacion'] = df2['identificacion'].astype(str)
df2['fec_matricula'] = pd.to_datetime(df2['fec_matricula'])
df2['fec_renovacion'] = pd.to_datetime(df2['fec_renovacion'])
df2['ult_ano_ren'] = df2['ult_ano_ren'].astype(str)
df2['mun_comercial'] = df2['mun_comercial'].astype(str)
df2['personal'] = df2['personal'].astype(int)
```

```python
# revisar df2
pd.set_option('display.max_columns', None)
df2
```

### Cagar data en Redshift

```python
# establecer los parámetros de conexión
dbname = 'data-engineer-database'
host = 'xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx'
port = 'xxxx'
user = 'xxxxxxxxxx'
password = 'xxxxxxx'

# crear la conexión
conn = psycopg2.connect(
    dbname=dbname,
    host=host,
    port=port,
    user=user,
    password=password
)

# crear un cursor
cursor = conn.cursor()

# crear tabla en Redshift si no existe
create_table_query = '''
    CREATE TABLE IF NOT EXISTS claudio_caniullan_coderhouse.empleados (
        matricula         varchar(80),
        organizacion      varchar(100),
        razon_social      varchar(100),
        identificacion    varchar(80),
        fec_matricula     datetime,
        fec_renovacion    datetime,
        ult_ano_ren       varchar(80),
        mun_comercial     varchar(80),
        personal          int
    );
'''
cursor.execute(create_table_query)
conn.commit()

# cargar los datos del dataFrame df2 en la tabla empleados
dataframe = df2

# iterar sobre las filas del dataFrame y ejecutar una consulta INSERT para cada fila
for _, row in dataframe.iterrows():
    insert_query = '''
        INSERT INTO claudio_caniullan_coderhouse.empleados (matricula,organizacion,razon_social,identificacion,fec_matricula,fec_renovacion,ult_ano_ren,mun_comercial,personal)
        VALUES (%s, %s, %s, %s, %s, %s,%s, %s, %s);
    '''
    cursor.execute(insert_query, tuple(row))
    conn.commit()

# cerrar el cursor y la conexión
cursor.close()
conn.close()
```
